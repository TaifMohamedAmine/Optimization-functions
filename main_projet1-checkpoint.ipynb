1 - One dimentional optimization methods:

The use of the following one dimentional optimization methods is mainly for real function with only one variable, we'll be testing the said functions and then visualising them in a 2d plot.

the test function is the given :

                       f(x) = 0.65 - 0.75/(1+x**2) - 0.65*x*arctan(1/x) 

```python
#testing function :

def f(x):
    
    return 0.65 - 0.75/(1+x**2) - 0.65*x*np.arctan(1/x)

def fp(x): # first derivative of the function f
    
    return 1.5*x/(1+x**2)**2 + 0.65*x/(1+x**2) - 0.65 * np.arctan(1/x)

def fpp(x): # second derivative of the function f
    
    return 1.5*(1-3*x**2)/(1+x**2)**3 + 0.65*(1-x**2)/(1+x**2)**2 + 0.65/(1+x**2)
```

testing the algorithms :

```python
%matplotlib inline  

X = np.linspace(0.1, 2, 3000)

f2 = np.vectorize(f)

print("*"*50)

# fixed step size : 

X1 = opt.fixed_step(f, 0.1, 0.00001 )

print("the minimum following the fixed step size method is : " ,X1)

plt.plot(X, f2(X), color = 'green')

plt.axis([0.1 ,1, -0.33, 0.1])

plt.scatter(X1, f(X1))

plt.title(' fixed step size ')

plt.show()

print("*"*50)

#accelerated step size :

X2 = opt.accelerated_step(f, 0.1, 0.00000000001, 0.00000001)

print("the minimum following the accelerated step size method is : " ,X2)

plt.plot(X, f2(X), color = 'red')

plt.axis([0.1 ,1, -0.33, 0.1])

plt.scatter(X2, f(X2))

plt.title(' accelerated step size ')

plt.show()

print("*"*50)

#halving method : 

X3 = opt.halving(f, 0, 2.5, 0.0000001)

print("the minimum following the interval halving method is : " ,X3)

plt.plot(X, f2(X), color = 'cyan')

plt.axis([0.1 ,1, -0.33, 0.1])

plt.scatter(X3, f(X3))

plt.title(' inteval halving ')

plt.show()

print("*"*50)

#secant search method :
      
X4 = opt.secant_Method(fp,0.01, 0.001)
      
print("the minimum following the secant search method is :", X4)

plt.plot(X, f2(X), color = 'yellow')

plt.axis([0.1 ,1, -0.33, 0.1])

plt.scatter(X4, f(X4))

plt.title('Secant Method ')

plt.show()

print("*"*50)

#dichotomous search method :

X5 = opt.dichotomous(f, 0, 2.5,0.0000001)

print("the minimum following the dichotomous method is :  " ,X5)

plt.plot(X, f2(X), color = 'magenta')

plt.axis([0.1 ,1, -0.33, 0.1])

plt.scatter(X5, f(X5))

plt.title(' dichotomous method ')

plt.show()
      
print("*"*50)

#fibonacci search method:
      
X6 = opt.fibonacci_search(f, 0.1, 2.5, 0.0000001, 100)
      
print("the minimum following the fibonacci search method is : " ,X6)

plt.plot(X, f2(X), color = 'black')

plt.axis([0.1 ,1, -0.33, 0.1])

plt.scatter(X6, f(X6))

plt.title(' Fibonacci Method ')

plt.show()
      
print("*"*50)
      
#Golden ratio method :
    
X7 = opt.golden_section(f, 0.1, 2.5, 0.0000001)
    
print("the minimum following the golden section method is : " ,X7)

plt.plot(X, f2(X))

plt.axis([0.1 ,1, -0.33, 0.1])

plt.scatter(X7, f(X7))

plt.title(' Golden section method ')

plt.show()
      
print("*"*50)
      
#Newton Rapson method :
    
X8 = opt.Newton_Rapson(fp, fpp , 0.1, 0.0000001)
      
print("the minimum following the Newton rapson method is :", X8)

plt.plot(X, f2(X), color = 'green')

plt.axis([0.1 ,1, -0.33, 0.1])

plt.scatter(X8, f(X8))

plt.title(' Newton Rapson ')

plt.show()
      
print("*"*50)
      
#Quasi Newton method :
      
X9 = opt.quasi_Newton_1D(f, 0.1, 0.01, 0.0000001)
      
print("the minimum following the Quasi newton method is :", X9)

plt.plot(X, f2(X), color = 'red')

plt.axis([0.1 ,1, -0.33, 0.1])

plt.scatter(X9, f(X9))

plt.title(' Quasi Newton ')

plt.show()
      
print("*"*50)

#exhaustive method :

X10 = opt.exhaustive(f, 0.1, 2.5, 10000)

print("the minimum following the exhaustive search method is : " ,X10)

plt.plot(X, f2(X), color = 'cyan')

plt.axis([0.1 ,1, -0.33, 0.1])

plt.scatter(X10, f(X10))

plt.title('exhaustive method ')

plt.show()

print("*"*50)
      
```

2 - Solving a linear system :

Solving a linear system is an important tool in most optimization problems. In this semester we learned three different methods to solve a linear system : Gauss, Choleski decomposition and LU decomposition. 

The linear system to be solved is supposed given in the format : 

                                             Ax = b
                                             
Where A is a matrix and b a vector of the same size as the number of A's columns.  

Our objective is to determine X for the said equation. 

2.1.1 - inverse of a function : 

```python
#testing the inverse of a matrix

P = np.array([[1,2,1],[1,2,2],[1,0,1]],dtype = float)

print("the matrix P before the transformation :")

print(P)

P_inverse, I_result = opt.Inverse_Matrice(P)

print("The matrix P after the transformation :")

print(I_result)

print("The inverse of the matrix is :")

print(P_inverse)

#testing the result

C = P_inverse @ P

print("the multiplication of P and its inverse matrix is :")

print(C)
```

2.1.2 - Choleski decomposition :

```python
#testing Choleski decomposition

A = np.array([[25,15,-5],[15,18,0],[-5,0,11]],dtype = float) 

L = opt.choleski(A)

# *** test ***

if np.all(opt.choleski(A) != None):
    print("The matrix L of the decomposition :")
    print(L)
    print("multiplication test :")
    C = L @ (np.transpose(L))
    test = np.all(A == C)
    if test == True :
        print("The decomposition is correcte : A = L @ L.T =")
        print(C)
else:
    print("la matrice n'est pas definie positive")

```

```python
import scipy.optimize as spo

#iterations of the Gradient descent on a contour : 

def g(X):
    return np.sin(X[0]) + np.cos(X[1]*X[0]) + 0.5*(X[0]**2 + 2*X[1]**2)

def g2(x,y):
    return np.sin(x) + np.cos(x*y) + 0.5*(x**2 + 2*y**2)

X_value = np.linspace(-3, 3 , 100)
Y_value = np.linspace(-3, 3, 100)

X, Y= np.meshgrid(X_value, Y_value)

Z = g2(X,Y)

fig = plt.figure()

plt.contour(X, Y, Z,10)


def gradient_descent_plot(f,x0, delta=0.01):
    X = x0
    nrm = np.linalg.norm
    grad = Gradient(f)
    while (nrm(grad(X)) > delta):
        phi = lambda alpha : f(X - alpha * grad(X))
        alpha_k = opt.accelerated_step(phi, 0, 0.0001, 0.001)
        plt.scatter(X[0], X[1])
        E = X
        X = X - alpha_k*grad(X)
        U = X
        plt.scatter(X[0], X[1])
        plt.annotate("", (U[0], U[1]), (E[0], E[1]), arrowprops={'arrowstyle': '->', 'color': 'r', 'lw': 1},
                 va='center', ha='center')
    return X

v = gradient_descent_plot(g , [0, 0])

plt.title("iterations of the Gradient descent")

plt.show()



```

```python
from numpy import linalg as alg
from numdifftools import Gradient, Hessian
import numpy as np


#iterations of the Gradient descent on a contour : 

def g(X):
    return np.sin(X[0]) + np.cos(X[1]*X[0]) + 0.5*(X[0]**2 + 2*X[1]**2)

def g2(x,y):
    return np.sin(x) + np.cos(x*y) + 0.5*(x**2 + 2*y**2)

X_value = np.linspace(-3, 3 , 100)
Y_value = np.linspace(-3, 3, 100)

X, Y= np.meshgrid(X_value, Y_value)

Z = g2(X,Y)

fig = plt.figure()

plt.contour(X, Y, Z,10)


def Newton_Method_plot(f, X0, delta = 0.001):
    grd = Gradient(f)
    norm = alg.norm
    inv = alg.inv
    H = Hessian(f)
    n = len(X0)
    X = X0
    I = np.identity(n)
    d = -np.dot(inv(H(X)), grd(X))
    while (norm(d) > delta):
        phi = lambda alpha : f(X - alpha * d)
        alpha_k = opt.accelerated_step(phi, 0, 0.0001, 0.001)
        plt.scatter(X[0], X[1])
        E = X
        X = X - alpha_k*d
        print(X)
        U = X
        plt.scatter(X[0], X[1])
        plt.annotate("", (U[0], U[1]), (E[0], E[1]), arrowprops={'arrowstyle': '->', 'color': 'r', 'lw': 1},
                 va='center', ha='center')
        if (np.all(alg.eigvals(H(X))) > 0) :
            d = -np.dot(inv(H(X)), grd(X))
        else:
            d = -np.dot(inv(delta*I + H(X)), grd(X))
    return X


v = Newton_Method_plot(g , [0, 0])

plt.title("iterations of the Newton method")

plt.show()
```

3.3 - Conjugate Gradient :

This method is different than the previous two algorithms as it requires a matrix Q and a vector b to write the function and it goes :

                f(X) = 1/2[X.T @ Q @ X] - b.T @ X

```python
import numpy as np

#testing function : (example taken from the lesson document )

Q = np.array([[2/3,1/5],[1/5,2/3]])
b_Q = np.array([0,0])
X0 = np.array([-0.5, 2])


def f_conjugate(x):
    return 1/2 * (np.transpose(x))@Q@x - (np.transpose(b_Q))@x

# importing the Gradient conjugate function from our optimization module

v3 = opt.Gradient_conjugate(f_conjugate,X0, Q )

print("the minimum point of the function following the Gradient conjugate is :")

print(v3)
```

```python
import numpy as np
from numdifftools import Gradient 

#iterations of the Gradient descent on a contour : 

Q = np.array([[2/3,1/5],[1/5,2/3]])
b_Q = np.array([0,0])
X0 = np.array([-0.5, 2])


def f_conjugate(x):
    
    return 1/2 * (np.transpose(x))@Q@x - (np.transpose(b_Q))@x

def f_conjugate2(x, y):
    
    return 1/2 * ((2/3 * x**2) + (2/5 * x*y) + (2/3 * y**2))


X_value = np.linspace(-3, 3 , 100)
Y_value = np.linspace(-3, 3, 100)

X, Y= np.meshgrid(X_value, Y_value)

Z = f_conjugate2(X,Y)

fig = plt.figure()

plt.contour(X, Y, Z,10)


def Gradient_conjugate(f, X0, Q ):
    n = len(X0)
    grd = Gradient(f)
    d = -grd(X0)
    X = X0
    print(X)
    for k in range(0, n):
        alpha_k = (np.transpose(d)@d)/(np.transpose(d)@Q@d)
        plt.scatter(X[0], X[1])
        Z = X
        X = X + alpha_k*d
        H = X
        plt.scatter(X[0], X[1])
        print(X)
        plt.annotate("", (H[0], H[1]), (Z[0], Z[1]), arrowprops={'arrowstyle': '->', 'color': 'r', 'lw': 1},
                 va='center', ha='center')
        beta_k = (np.transpose(grd(X))@Q@d)/(np.transpose(d)@Q@d)
        d = -grd(X) + beta_k*d
    return X

v3 = Gradient_conjugate(f_conjugate,X0, Q )



plt.scatter(v3[0], v3[1])


plt.title("iterations of the Gradient conjugate method")

plt.show()
```
